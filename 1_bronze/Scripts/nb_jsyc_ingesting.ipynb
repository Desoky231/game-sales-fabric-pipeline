```python
file_path = 'Files/raw/games'
table_name = 'games'
```

```python
import re
import json
from datetime import datetime
```

```python
try:
    df = spark.read.format("parquet").load(file_path)
    rows_in = df.count()

    cleaned_columns = [re.sub(r'[^\w]', '_', col.strip()).lower() for col in df.columns]
    df_clean = df.toDF(*cleaned_columns)

    rows_out = df_clean.count()

    df_clean.write.mode("overwrite") \
        .option("overwriteSchema", "true") \
        .format("delta") \
        .saveAsTable(table_name)

    status = "Succeeded"

except Exception as e:
    print(f"Error during ingestion: {e}")

finally:
    output = {
        "rows_in": rows_in,
        "rows_out": rows_out,
        "status": status,
        "run_ts": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
```

```python
notebookutils.notebook.exit(json.dumps(output))
```

```python
from pyspark.sql.functions import col
```

```python
promotions_dim = spark.sql("SELECT * FROM lh_silver.promotions")
```

```python
promotions_dim = promotions_dim.select(col("promotion_id") , col("name"), col("start_date") , col("end_date"))
```

```python
promotions_dim.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('promotions_dim')
```

```python
from pyspark.sql import functions as F
```

```python
wishlist = spark.sql("SELECT * FROM lh_silver.wishlist")
```

```python
wishlist_with_date_id = wishlist.withColumn(
    'wishlist_date_id',
    F.date_format('wishlist_date', 'yyyyMMdd').cast('int') 
)
```

```python
wishlist_with_date_id.printSchema()
```

```python
from pyspark.sql.functions import col

wishlist_with_date_id = wishlist_with_date_id.withColumn(
    "user_id", col("user_id").cast("long")
)

wishlist_with_date_id.printSchema()
```

```python
wishlist_final = wishlist_with_date_id.drop('wishlist_date')
```

```python
wishlist_final.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable("wishlist_fact")
```

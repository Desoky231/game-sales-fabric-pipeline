```python
from pyspark.sql.functions import col
```

```python
games_dim = spark.sql("SELECT * FROM lh_silver.games")
```

```python
games_categories = spark.sql("SELECT * FROM lh_silver.games_categories")
```

```python
games_categories_bridge = spark.sql("SELECT * FROM lh_silver.games_categories_bridge")
```

```python
games_categories_final = games_categories.join(games_categories_bridge, [games_categories_bridge.categories_id == games_categories.id]).select(col('appid'), col('categories'))
```

```python
display(games_categories_final.orderBy('appid'))
```

```python
games_developers = spark.sql("SELECT * FROM lh_silver.games_developers LIMIT 1000")
```

```python
games_developers_bridge = spark.sql("SELECT * FROM lh_silver.games_developers_bridge LIMIT 1000")
```

```python
games_developers_bridge.printSchema()
```

```python
games_developers_final = games_developers.join(games_developers_bridge, [games_developers_bridge.developers_id == games_developers.id]).select(col('appid'), col('developers'))
```

```python
games_publishers = spark.sql("SELECT * FROM lh_silver.games_publishers")
```

```python
games_publishers_bridge = spark.sql("SELECT * FROM lh_silver.games_publishers_bridge")
```

```python
games_publishers_final = games_publishers.join(games_publishers_bridge, [games_publishers_bridge.publishers_id == games_publishers.id]).select(col('appid'), col('publishers'))
```

```python
games_genres = spark.sql("SELECT * FROM lh_silver.games_genres")
```

```python
games_genres_bridge = spark.sql("SELECT * FROM lh_silver.games_genres_bridge")
```

```python
games_genres_final = games_genres.join(games_genres_bridge, [games_genres_bridge.genres_id == games_genres.id]).select(col('appid'), col('genres'))
```

```python
games_dim.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('games_dim')
```

```python
games_genres_final.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('genre_lookup')
```

```python
games_publishers_final.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('publisher_lookup')
```

```python
games_developers_final.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('developer_lookup')
```

```python
games_categories_final.write.mode('overwrite').option("overwriteSchema", "true").format("delta").saveAsTable('category_lookup')
```
